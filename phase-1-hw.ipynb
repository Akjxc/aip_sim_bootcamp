```python
import pandas as pd

filepath = '/kaggle/input/superstore-dataset-final/Sample - Superstore.csv'
df = pd.read_csv(filepath, index_col=0, encoding='latin1')
df
```

카테고리별로 총 판매량, 총 수익, 평균 할인율을 비교하라.

각 고객이 몇 번이나 주문했는지와, 그들의 평균 구매 금액은 얼마인지 확인하라.

```python
# 2.
df
# Customer ID, Customer Name 중복됨
# Product ID, Product Name 중복됨
result = df.drop(columns=['Customer Name', 'Product Name'])
result
```

```python
# 3.
def cal_arrived(order, ship):
    order = pd.to_datetime(order)
    ship = pd.to_datetime(ship)
    
    df['Duration'] = ship - order
    return df
    
cal_arrived(df['Order Date'], df['Ship Date'])
df
```

```python
# 4.

# total_sales = df.groupby('Category').Sales.sum()
# discount_mean = df.groupby('Category').Discount.mean()
# total_profit = df.groupby('Category').Profit.sum()

# table = pd.DataFrame({'Total Sales': total_sales, 'Discount Mean': discount_mean, 'Total Profit': total_profit}, index=df['Category'].unique())
# table

df.groupby('Category').agg({'Sales': 'sum', 'Discount': 'mean', 'Profit': 'sum'}).rename(columns={'Sales': 'Total Sales', 'Discount': 'Discount Mean', 'Profit': 'Total Profit'})
```

```python
# 5. 각 고객이 몇 번이나 주문했는지와, 그들의 평균 구매 금액은 얼마인지 확인하라.
customer_sales_mean = df.groupby('Customer ID').agg({'Order ID': 'count', 'Sales': 'mean'}) 
customer_sales_mean.rename(columns={'Order ID': 'Count', 'Sales': 'Sales Mean'})
# agg - 단일 프로퍼티의 다중 연산 : 리스트 사용 / 다중 프로퍼티 각각 연산 : 딕셔너리
# df.groupby('Customer ID')['Sales'].mean()
```


✅ EDA / 시각화 준비

지역(Region)과 고객 유형(Segment)별로 매출과 수익률을 비교하라.

월별 매출 추이를 분석하기 위한 테이블을 생성하라.

수익률이 가장 높은 상품 5개를 찾아라. (수익률: 수익 ÷ 매출)

```python
# 1.지역(Region)과 고객 유형(Segment)별로 매출과 수익률을 비교하라.
df.groupby(['Region', 'Segment']).agg({'Sales': 'sum', 'Profit': 'sum'})
```

```python
# 2. 월별 매출 추이를 분석하기 위한 테이블을 생성하라.
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Month'] = df['Order Date'].dt.month
# agg - 단일함수 사용 : Series, 다중함수 사용 : DataFrame 반환 / 
# 단일함수 사용하면서 DataFrame 뽑고싶다 - 리스트 감싸기  ex) agg(['sum'])
table = df.groupby('Month').Sales.sum().reset_index().set_index('Month')
# df.groupby('~~').property.sum() - Series
# 거기에 reset_index - DataFrame 반환
# reset_index: Series의 index를 컬럼으로 끌어내리고 새로운 인덱스 부여.
table
```

```python
# 3. 수익률이 가장 높은 상품 5개를 찾아라. (수익률: 수익 ÷ 매출)
product_total_sales = df.groupby('Product ID').Sales.sum()
product_total_profit = df.groupby('Product ID').Profit.sum()
return_rate = (product_total_profit / product_total_sales)
return_rate.sort_values(ascending=False).head(5)
```

✅ 선택 문제 (심화 분석)

배송에 평균보다 오래 걸린 주문을 찾아라.

가장 높은 할인율로 판매된 주문은 어떤 제품이었고, 수익은 얼마나 발생했는가?

```python
# 1. 배송에 평균보다 오래 걸린 주문을 찾아라.
df['Order ID'].loc[df.Duration > df.Duration.mean()]
```

```python
# 2. 가장 높은 할인율로 판매된 주문은 어떤 제품이었고, 수익은 얼마나 발생했는가?
discount_max = df.Discount.max()
df[['Product ID', 'Profit', 'Discount']].loc[df.Discount == discount_max]
# 여러 컬럼 select는 리스트에 감쌀 것.
```
