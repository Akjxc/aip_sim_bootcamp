ü•â Í≤∞Ï∏°Ïπò/ÌÉÄÏûÖÎ≥ÄÌôò (2Î¨∏Ï†ú)
1. "sales" Ïª¨ÎüºÏùò Í≤∞Ï∏°ÏπòÎ•º sales ÌèâÍ∑†Í∞íÏúºÎ°ú Ï±ÑÏõåÎùº.

2. "signup_date" Ïª¨ÎüºÏùÑ datetime ÌÉÄÏûÖÏúºÎ°ú Î≥ÄÌôòÌï¥Îùº.

    - Í∑∏Î¶¨Í≥† year, month, day Ïª¨ÎüºÏùÑ Í∞ÅÍ∞Å ÏÉàÎ°ú ÎßåÎì§Ïñ¥ Ï∂îÍ∞Ä

```python
def missing_sales(row):
    if pd.isna(row.sales):
        row.sales = df.sales.mean()
    return row

df.apply(missing_sales, axis='columns')
```

```python
df['signup_date'] = pd.to_datetime(df['signup_date'])
datetime = pd.DataFrame({'year': df['signup_date'].dt.year, 'month': df['signup_date'].dt.month, 'day': df['signup_date'].dt.day})

df.join(datetime)
```

üèÖ Í≥†Í∏â Ï°∞Ïûë (3Î¨∏Ï†ú)
1. Í≥†Í∞ùÎ≥Ñ ÎàÑÏ†Å Íµ¨Îß§ Í∏àÏï° Ïª¨Îüº cumulative_salesÎ•º ÎßåÎì§Ïñ¥Îùº.

    - sort_values(['customer_id', 'date']) ÌõÑ groupby ‚Üí cumsum Ï†ÅÏö©

2. pivot_tableÏùÑ ÎßåÎì§Ïñ¥Îùº.

    - Ìñâ: region

    - Ïó¥: product_category

    - Í∞í: salesÏùò Ìï©Í≥Ñ

3. Ï°∞Í±¥Î∂Ä ÏÉàÎ°úÏö¥ Ïª¨Îüº ÎßåÎì§Í∏∞

    - sales >= 10000Ïù¥Î©¥ 'High', ÏïÑÎãàÎ©¥ 'Low'Î•º Í∏∞Î°ùÌïòÎäî sales_level Ïª¨Îüº Ï∂îÍ∞Ä


```python
# cumulative_sales = df.sort_values(['customer_id', 'date']).groupby('customer_id').sales.cumsum()
# cumulative_sales

# 1) ÎÇ†Ïßú ÏàúÏÑúÎåÄÎ°ú Ï†ïÎ†¨
df_sorted = df.sort_values(['customer_id', 'date'])

# 2) Í≥†Í∞ùÎ≥ÑÎ°ú sales ÎàÑÏ†ÅÌï© Í≥ÑÏÇ∞
running = df_sorted.groupby('customer_id')['sales'].cumsum()

# 3) ÏõêÎ≥∏(ÎòêÎäî Ï†ïÎ†¨Îêú) dfÏóê Ïª¨Îüº Ï∂îÍ∞Ä
df_sorted['cumulative_sales'] = running

df_sorted
```

```python
df_pivot = pd.pivot_table(data=df, index='region', columns='product_category', values='sales', aggfunc='sum', fill_value=0)
grouped_df = df.groupby(['region', 'product_category']).sales.sum()
grouped_df
df_pivot
```

ü•á Ïã§Ï†Ñ Ï¢ÖÌï© Î¨∏Ï†ú (1Î¨∏Ï†ú)
1. ÏïÑÎûò ÏûëÏóÖÏùÑ ÏàúÏÑúÎåÄÎ°ú ÏßÑÌñâÌïòÎäî Ìï®Ïàò process_sales(df)Î•º ÎßåÎì§Ïñ¥Îùº.

    - Í≤∞Ï∏°Ïπò Ï≤òÎ¶¨ (salesÎäî ÌèâÍ∑†ÏúºÎ°ú)

    - 'signup_date'Î•º datetime Î≥ÄÌôòÌïòÍ≥† year Ï∂îÏ∂ú

    - sales_level (High/Low) Ï∂îÍ∞Ä - 10000 Í∏∞Ï§Ä

    - ÏµúÏ¢ÖÏ†ÅÏúºÎ°ú customer_id, sales, sales_level, year Ïª¨ÎüºÎßå ÎÇ®Í∏¥ DataFrame Î∞òÌôò
  

```python
def sales_judge(df):
    if df.sales > 10000:
        return 'High'
    else:
        return 'Low'
        

def process_sales(df):
    df.sales = df.sales.fillna(df.sales.mean()) # Í≤∞Ï∏°Ïπò Ï≤òÎ¶¨
    
    df['signup_date'] = pd.to_datetime(df['signup_date']) # datetime Î≥ÄÌôò, year Ï∂îÏ∂ú
    df['year'] = df['signup_date'].dt.year

    df['sales_level'] = df.apply(sales_judge, axis='columns')
    
    return df[['customer_id', 'sales', 'sales_level', 'year']]

process_sales(df)
    
    
```
